## Exploration Parquet avec logs
Ajout d'une exploration orientee parquet avec journalisation des etapes.

```python
import logging
import polars as pl
from pathlib import Path

logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)s | %(message)s")
logger = logging.getLogger("exploration_parquet")

PARQUET_PATH = Path("./DATA/complete_dataset.parquet")
logger.info("Lecture du fichier: %s", PARQUET_PATH)

lf = pl.scan_parquet(PARQUET_PATH)
schema = lf.collect_schema().names()
stats = lf.select([pl.len().alias("rows"), pl.n_unique("siret").alias("unique_siret")]).collect()
logger.info("Statistiques globales: %s", stats.to_dicts()[0])

pl.DataFrame({"columns": schema})
```

```python
logger.info("Calcul des taux de valeurs manquantes")
null_profile = (
    lf.null_count()
    .collect()
    .transpose(include_header=True, header_name="column", column_names=["null_count"])
    .with_columns((pl.col("null_count") / stats["rows"][0] * 100).round(2).alias("null_pct"))
    .sort("null_pct", descending=True)
    .head(20)
)
null_profile
```
